```python
from matplotlib import pyplot as plt
import gym, gym_unbalanced_disk, time
import numpy as np
import pickle


with open('sim_data_ob', 'rb') as f:
    observations = pickle.load(f)

with open('sim_data_reward', 'rb') as f:
    observations_sim = pickle.load(f)

plt.figure(figsize=(14,8))
plt.plot(observations, label=['theta','omega'])
plt.xlabel('Rotation angle of the disc',fontsize=12)
plt.ylabel('y',fontsize=14)
plt.legend(fontsize=14)
```

```python
theta = []
omega = []
for i in range(len(observations)):
    theta.append((observations[i][0]-9)/10*np.pi)
    omega.append(observations[i][1]-9)
```

```python
plt.plot(np.abs(theta),label='$|\\theta|$')
plt.plot(np.abs(omega),label='$|\omega|$')
plt.plot([np.pi]*len(theta),'--',label='$\pi$')
plt.xlabel('Iterations')
plt.ylabel('Theta')
plt.legend()
plt.show()
plt.plot(observations_sim, label='f = $-10*\\theta^2 - 0.1*\omega^2 - 1e^{-5}*u^2$')
plt.xlabel('Iterations')
plt.ylabel('Rewards')
plt.legend()
```
